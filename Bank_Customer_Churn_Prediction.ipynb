```python
#importing dependencies
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#importing dataset
dataset = pd.read_csv('/content/Churn_Modelling.csv')
```

```python
dataset.head()
```

```python
#getting some info about the dataset
dataset.info()
```

```python
#cheking the null values
dataset.isnull().sum()
```

```python
dataset['Exited'].value_counts()
```

```python
dataset = dataset.drop(['RowNumber','CustomerId','Surname'], axis=1)
```

```python
dataset['Geography'] = dataset['Geography'].map({'France': 1, 'Germany': 0, 'Spain': 2})
```

```python
dataset['Gender'] = dataset['Gender'].map({'Male': 1, 'Female': 0})
```

```python
X = dataset.drop(columns='Exited',axis=1)
Y = dataset['Exited']
```

```python
X
```

```python
Y
```

```python
#splitting the data into traintestsplit
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=4)
```

```python
X.shape,X_train.shape,X_test.shape
```

```python
scaler = StandardScaler()
```

```python
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
#Building an Ann model
model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(10,)))
model.add(keras.layers.Dense(units=10,activation='relu'))
model.add(keras.layers.Dense(units=15,activation='relu'))
model.add(keras.layers.Dense(2,activation='sigmoid'))
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(X_train_std,Y_train,validation_split=0.1,epochs=30)
```

```python
loss,accuracy = model.evaluate(X_test_std,Y_test)
print("Accuracy for testing data is ",round(accuracy*100,2),"%")
```
